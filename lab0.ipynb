```python
'''
new material before problem 1


Usually, to loop through a list called xs,
we would do
'''
for x in xs:
    print(x)
    
'''
But sometimes we want to keep track
of an explicit loop index, call it i.
We can first check the length of a list like this:
'''
print("length of list xs is:",len(xs))

'''
Then, we can recall that  range(n) behaves like
a list [0,1,2,3,...,n-1] when using it 
as the list in a for-loop.

Finally, we can use len() in within range() 
to loop through a list of indexes,
rather than elements of the list xs themselves.
Then, we can index into xs with the indexes.
'''
for i in range(len(xs)):
    print("ith index of xs is:",xs[i])    
'''

check if a number is even:

In math, the modulus operator works like this:
0 mod 3 is 0
1 mod 3 is 1
2 mod 3 is 2
3 mod 3 is 0
4 mod 3 is 1
5 mod 3 is 2
6 mod 3 is 0
etc...

That means, to check if a number x is even,
we should check whether x mod 2 is equal to 0.

In python, mod is denoted by the percent symbol %
'''
x = 4
print("x mod 2 is:",x % 2)
```

```python
'''
New material before problem 4:

Sometimes we need to check whether
several conditions are True simultaneously.
We can always connect True or False values
logically with operators "and" (called a conjunction) 
and "or" (called a disjunction).
These operators satisfy the following rules:

if A is True and B is True, then their conjunction
A AND B is also true. Otherwise, if either or both
of A,B are False, then A AND B is false.
'''
A = True
B = True
print(A and B)
'''
in python we use special keyword "and" for the conjunction AND.
in other languages, you may see "&&". 
try changing the value of A and B and make sure
conjunctions follows the description above
'''

'''
Alternatively, a disjunction A OR B
is True as long as either A is True or B is True.
This means A OR B is True when either or both are True,
and it's false only when both A is False and B is False
'''
A = True
B = False
print(A or B)

'''
in python we use special keyword "or" for the disjunction.
in other languages, you may see "||".
try changing the value of A and B and make sure
disjunctions follows the description above.
'''
```
